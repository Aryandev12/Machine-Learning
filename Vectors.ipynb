```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  #data visualization library.
sns.set()
```

Plotting a vector

```python
plt.quiver(0,0,4,5)
plt.show()
```

```python
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1)
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#color property
plt.quiver(0,0,4,5,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,5,scale_units='xy',angles='xy',scale=1,color='y')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Addition of two vectors.

```python
vector_1 = np.asarray([0,0,2,3])
vector_2 = np.asarray([0,0,3,-2])
sum = vector_1+ vector_2 #for substraction do vector_1 - vector_2
print(sum)
a,b,c,d = sum
```

```python
#color property
plt.quiver(0,0,2,3,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,3,-2,scale_units='xy',angles='xy',scale=1,color='y')
plt.quiver(a,b,c,d,scale_units='xy',angles='xy',scale=1,color='g')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
